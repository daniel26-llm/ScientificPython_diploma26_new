```python
from scipy.interpolate import interp1d


# generate data
x = np.linspace(0, 10, 17)
y = np.cos(-x**2/9.0)

# interpolate
f = interp1d(x, y, kind='cubic')
x_interp = np.linspace(0, 10, num=51)
y_interp = f(x_interp)

# plot original and interpolated data
plt.plot(x, y, '--', label='Original data')
plt.plot(x_interp, y_interp, label='Interpolated data')
plt.legend()
```

```python
#exercise7 
from scipy.interpolate import interp1d


# generate data
x = np.linspace(0, 10, 17)
y = np.cos(-x**2/9.0)

# interpolate
f = interp1d(x, y, kind='nearest')
x_interp = np.linspace(0, 10, num=51)
y_interp = f(x_interp)

# plot original and interpolated data
plt.xlabel('Random Number Generated')
plt.ylabel('cos(x)')
plt.title('Data Fitting With SciPy using nearest')
plt.plot(x, y, '--', label='Original data')
plt.plot(x_interp, y_interp, label='Interpolated data')
plt.legend()
```

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0,5,20)
y1 = np.sin(x)
y2 = np.cos(x + 0.2)
y3 = np.tan(x + 0.4)
y4 = np.sinh(x + 0.6)
y5 = np.cosh(x + 0.8)
y6 = np.tanh(x + 1)

# now we decide the actual figure size in inches
fig = plt.figure(figsize=(14,7))

plt.suptitle('A graph of trigonometric functions and Hyperbolic Functions',fontweight='bold', fontsize='x-large')
plt.subplots_adjust(hspace=0.3, top=0.8)

# create subplots 231 means make a 2x3 grid and this is the first plot
plt.subplot(231)
#these are here to show that you can do the same here as for a single plot
plt.title('Random number against sin(x)',fontsize = 'large',fontstyle='italic',fontweight='bold')
plt.ylabel('sin(x)',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('random numbers',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.plot(x,y1,color='r')

plt.subplot(232)
plt.title('cos function')
plt.xlabel('random numbers')
plt.ylabel('cos(x)')
plt.plot(x,y2,color='g')

plt.subplot(233)
plt.title('tan function')
plt.xlabel('random numbers')
plt.ylabel('tan(x)')
plt.plot(x,y3,color='b')

plt.subplot(234)
plt.title('sinh function')
plt.xlabel('random numbers')
plt.ylabel('sinh(x)')
plt.plot(x,y4,color='y')

plt.subplot(235)
plt.title('cosh function')
plt.xlabel('random numbers')
plt.ylabel('cosh(x)')
plt.plot(x,y5,color='c')

plt.subplot(236)
plt.title('tanh function')
plt.xlabel('random numbers')
plt.ylabel('tanh(x)')
plt.plot(x,y6,color='m')
# removed extra white space
plt.tight_layout()
#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


R = 2      
r = 0.7    


u = np.linspace(0, 2*np.pi, 200)
v = np.linspace(0, 2*np.pi, 200)
u, v = np.meshgrid(u, v)


x = (R + r*np.cos(v)) * np.cos(u)
y = (R + r*np.cos(v)) * np.sin(u)
z = r * np.sin(v)


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(
    x, y, z,
    rstride=4, cstride=4,
    cmap='viridis',
    edgecolor='none',
    alpha=0.95
)


ax.set_box_aspect([1,1,1])
ax.set_title("Smooth 3D Torus")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
#2 sequence of numbers 
x = [0]  

n = int(input('enter any number: '))
i = 1   
while i < n:
    cond1 = x[i-1] - i
    if cond1 > 0 and cond1 not in x:
        x.append(cond1) 
    else:
        x.append(x[i-1] + i)
    i += 1

#print(x) 
arr1 = np.array(x)
arr2 = np.linspace(0,100,100)
plt.title('The recaman series')
plt.xlabel('Counting numbers from 0-100')
plt.ylabel('Recaman series for each value of N')
plt.plot(arr2, arr1) 
plt.legend()
plt.show()
```

```python
#matrix 
def mat_mult(A, B):
    if len(A[0]) != len(B):
        print('No possible!')
        
    return [[sum(A[i][k] * B[k][j] for k in range (len(B)))
            for j in range(len(B[0]))]
            for i in  range(len(A))]
A = [[1,2],[5,6],[3,0]]
B = [[1,4,6],[9,0,1]] 
mat_mult(A,B) 
```
