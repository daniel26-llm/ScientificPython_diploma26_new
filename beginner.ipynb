```python
#2 sequence of numbers 
x = [0]  

n = int(input('enter any number: '))
i = 1   
while i < n:
    cond1 = x[i-1] - i
    if cond1 > 0 and cond1 not in x:
        x.append(cond1) 
    else:
        x.append(x[i-1] + i)
    i += 1

print(x) 
```

```python
#q5
from math import isqrt 
def factor(n):
    if n <= 0:
        return []
    else:
        return [ (i, n//i) for i in range(1, isqrt(n)+1) if n % i == 0
               ]
factor(36) 
```

```python
#7
def sieve_of_eratosthenes(n):
    if n < 2:
        return []
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    p = 2
    while p * p <= n:
        if is_prime[p]:
            for multiple in range(p * p, n + 1, p):
                is_prime[multiple] = False
        p += 1
    return [i for i in range(n + 1) if is_prime[i]]
sieve_of_eratosthenes(10)  
```

```python
#8 palindrome 
def palindrome():
    a = str(input("write a word:"))
    b = a.upper()
    m = len(b)
    word = []

    for i in range(m):
        if b[i] == ' ':
            b.replace(' ','')
        else:
            word.append(b[i])

    print(word)
    reverse = word[::-1]
    print(reverse)

    if word == reverse:
        print("The word is a palindrome")
    else:
        print("The word is not a palindrome")
palindrome()
```
